```python
# Create Arrays.
import numpy as np

# From a list
a = np.array([1, 2, 3])
print(a)             # [1 2 3]

# 2D array (matrix)
b = np.array([[1, 2, 3], [4, 5, 6]])
print(b)
print("Shape:", b.shape)
```

```python
# Useful Array Constructors
z = np.zeros((2,3))       # 2x3 array of 0s
print("z:", z)

o = np.ones((2,3))        # 2x3 array of 1s
print("\no:", o)

e = np.eye(3)             # Identity matrix (3x3)
print("\ne:", e)

a = np.arange(0,10,2)     # [0 2 4 6 8]
print("\na:", a)

l = np.linspace(0,1,5)    # [0. 0.25 0.5 0.75 1.]
print("\nl:", l)
```

# Array Operations

```python
# Element-wise Math
a = np.array([1,2,3])
b = np.array([4,5,6])

print(a + b)   # [5 7 9]
print(a * b)   # [4 10 18]
# Raise every element in the array to the power of 2
# not the whole array as a single number, but each element individually.
print(a ** 2)  # [1 4 9]
```

```python
# Matrix Operations
A = np.array([[1,2],[3,4]])
B = np.array([[5,6],[7,8]])

print(A @ B)       # Matrix multiplication
print(np.dot(A, B))
```

### Visualize Matrix Multiplication.

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation

# --- Define Matrices ---
A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])

C = A @ B   # Matrix multiplication result

# --- Plot Setup ---
fig, axes = plt.subplots(1, 3, figsize=(9, 3))
titles = ['Matrix A', 'Matrix B', 'Result C = A @ B']
mats = [A, B, C]

for ax, title, mat in zip(axes, titles, mats):
    im = ax.imshow(mat, cmap='Blues', vmin=0, vmax=10)
    for (i, j), val in np.ndenumerate(mat):
        ax.text(j, i, f'{val}', ha='center', va='center', color='black', fontsize=12)
    ax.set_title(title)
    ax.set_xticks([])
    ax.set_yticks([])

fig.suptitle('Matrix Multiplication Visualization', fontsize=14)

# --- Animation ---
steps = [
    (0, 0, 0, 0),  # C11 = Row0•Col0
    (0, 0, 0, 1),  # C12 = Row0•Col1
    (1, 1, 0, 0),  # C21 = Row1•Col0
    (1, 1, 0, 1),  # C22 = Row1•Col1
]

def update(step):
    for ax in axes:
        ax.clear()
    for ax, title, mat in zip(axes, titles, mats):
        im = ax.imshow(mat, cmap='Blues', vmin=0, vmax=10)
        for (i, j), val in np.ndenumerate(mat):
            ax.text(j, i, f'{val}', ha='center', va='center', color='black', fontsize=12)
        ax.set_title(title)
        ax.set_xticks([])
        ax.set_yticks([])

    (r, rA, cB, c) = steps[step]
    # Highlight active row and column
    axes[0].axhline(r + 0.5, color='red', lw=2)
    axes[1].axvline(cB - 0.5, color='green', lw=2)
    axes[2].text(cB, r, f'{C[r, cB]}', ha='center', va='center', color='red', fontsize=14, fontweight='bold')

ani = animation.FuncAnimation(fig, update, frames=len(steps), interval=1000, repeat=True)
plt.show()
```

```python
# Array attributes.
arr = np.array([[1,2,3],[4,5,6]])
print(arr.shape)   # (2,3)
print(arr.ndim)    # 2
print(arr.size)    # 6
print(arr.dtype)   # int64 (depends on system)
```

```python
#Array Indexing and Slicing
arr = np.array([[10,20,30],
                [40,50,60],
                [70,80,90]])
print(arr.shape)

print(arr[0,1])   # 20
print(arr[:,1])   # [20 50 80]
print(arr[1,:])   # [40 50 60]
print(arr[0:2,1:3]) # [[20 30]
                    #  [50 60]]
```
