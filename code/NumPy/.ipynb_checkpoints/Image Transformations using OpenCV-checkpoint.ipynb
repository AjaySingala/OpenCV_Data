### Setup

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('people.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.title("Original Image")
plt.axis('off')
plt.show()
```

###  1. Translation (Move Image)

Move the image along X and Y axes.

```python
rows, cols = img.shape[:2]

# Move right by 100, down by 50
M = np.float32([[1, 0, 100],
                [0, 1, 50]])

translated = cv2.warpAffine(img, M, (cols, rows))

plt.imshow(translated)
plt.title("Translated Image")
plt.axis('off')
plt.show()
```

 2. Rotation

Rotate the image around its center.

```python
center = (cols // 2, rows // 2)
angle = 45      # degrees
scale = 1.0

M = cv2.getRotationMatrix2D(center, angle, scale)
rotated = cv2.warpAffine(img, M, (cols, rows))

plt.imshow(rotated)
plt.title("Rotated Image (45)")
plt.axis('off')
plt.show()
```

 3. Scaling (Resize / Zoom)

```python
# Shrink to half
small = cv2.resize(img, (0, 0), fx=0.5, fy=0.5)

# Enlarge by 1.5x
large = cv2.resize(img, (0, 0), fx=1.5, fy=1.5)

# Show
plt.subplot(1,2,1); plt.imshow(small); plt.title("Scaled Down")
plt.subplot(1,2,2); plt.imshow(large); plt.title("Scaled Up")
plt.show()
```

 4. Affine Transformation

Affine transformations keep lines parallel but can rotate, shear, or translate them.

```python
rows, cols = img.shape[:2]

# Three points before and after transformation
pts1 = np.float32([[50, 50], [200, 50], [50, 200]])
pts2 = np.float32([[10, 100], [200, 50], [100, 250]])

M = cv2.getAffineTransform(pts1, pts2)
affine = cv2.warpAffine(img, M, (cols, rows))

plt.imshow(affine)
plt.title("Affine Transformation")
plt.axis('off')
plt.show()
```

 5. Perspective Transformation

This lets you warp the image as if changing the camera view (useful in document scans or AR).

```python
rows, cols = img.shape[:2]

# Corners of original
pts1 = np.float32([[50, 50], [450, 50], [50, 300], [450, 300]])

# Where they should map to
pts2 = np.float32([[10, 100], [400, 50], [100, 300], [400, 300]])

M = cv2.getPerspectiveTransform(pts1, pts2)
perspective = cv2.warpPerspective(img, M, (cols, rows))

plt.imshow(perspective)
plt.title("Perspective Transformation")
plt.axis('off')
plt.show()
```
