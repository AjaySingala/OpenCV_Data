### Setup

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Load an image in grayscale
gray = cv2.imread('people.jpg', cv2.IMREAD_GRAYSCALE)

plt.imshow(gray, cmap='gray')
plt.title("Original Grayscale Image")
plt.axis('off')
plt.show()
```

###  Step 2: Apply a Color Map

OpenCV provides 22 built-in color maps under the cv2.COLORMAP_* constants.

```python
colored = cv2.applyColorMap(gray, cv2.COLORMAP_JET)

plt.imshow(cv2.cvtColor(colored, cv2.COLOR_BGR2RGB))
plt.title("Color Mapped Image (JET)")
plt.axis('off')
plt.show()
```

### Display Multiple Color Maps Side by Side

```python
color_maps = [
    cv2.COLORMAP_JET,
    cv2.COLORMAP_HOT,
    cv2.COLORMAP_BONE,
    cv2.COLORMAP_COOL,
    cv2.COLORMAP_INFERNO,
    cv2.COLORMAP_WINTER
]

plt.figure(figsize=(12,6))

for i, cmap in enumerate(color_maps):
    colored = cv2.applyColorMap(gray, cmap)
    colored = cv2.cvtColor(colored, cv2.COLOR_BGR2RGB)
    plt.subplot(2, 3, i+1)
    plt.imshow(colored)
    plt.title(str(cmap).split('_')[-1])
    plt.axis('off')

plt.tight_layout()
plt.show()
```

###  Step 4: Custom Color Mapping (Optional Advanced)

You can create your own color map using cv2.applyColorMap() with a custom LUT.

```python
# Create a custom LUT: black  red gradient
lut = np.zeros((256, 1, 3), dtype=np.uint8)
lut[:, 0, 2] = np.arange(256)  # Red channel increases

custom_colored = cv2.LUT(cv2.cvtColor(gray, cv2.COLOR_GRAY2BGR), lut)

plt.imshow(custom_colored)
plt.title("Custom Red Gradient")
plt.axis('off')
plt.show()
```
