```python
import cv2
from matplotlib import pyplot as plt

# Read image (by default OpenCV reads in BGR format)
img = cv2.imread('people.jpg')

# Convert BGR  RGB for correct color display in matplotlib
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Display
plt.imshow(img_rgb)
plt.title("RGB Image")
plt.axis('off')
plt.show()
```

```python
print("Shape:", img.shape)      # (height, width, channels)
print("Size:", img.size)        # Total number of pixels
print("Data type:", img.dtype)  # Data type (usually uint8)
```

```python
# Crop (ROI - Region of Interest)
crop = img[50:200, 100:300]

# Display
plt.imshow(crop)
plt.axis('off')
plt.show()
```

```python
# Resize
resized = cv2.resize(img, (300, 200))

# Display
plt.imshow(resized)
plt.axis('off')
plt.show()
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
hsv  = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
```

```python
# Rotate
rotated = cv2.rotate(img, cv2.ROTATE_90_CLOCKWISE)
#rotated = cv2.rotate(img, cv2.ROTATE_45_CLOCKWISE)

# Display
plt.imshow(rotated)
plt.axis('off')
plt.show()
```

```python
# Flip horizontally or vertically
flipped = cv2.flip(img, 1)  # 1 = horizontal, 0 = vertical

# Display
plt.imshow(flipped)
plt.axis('off')
plt.show()
```

```python
# Simple blur
blur = cv2.GaussianBlur(img, (5,5), 0)

# Display
plt.imshow(blur)
plt.axis('off')
plt.show()
```

```python
# Edge detection
edges = cv2.Canny(img, 100, 200)

# Display
plt.imshow(edges)
plt.axis('off')
plt.show()
```

```python
# Draw line, rectangle, circle, text
cv2.line(img, (50, 50), (200, 50), (0, 255, 0), 3)
cv2.rectangle(img, (60, 60), (180, 120), (255, 0, 0), 2)
cv2.circle(img, (120, 120), 30, (0, 0, 255), -1)
cv2.putText(img, "Hello", (50, 200), cv2.FONT_HERSHEY_SIMPLEX, 1, (255,255,255), 2)

# Display
plt.imshow(img)
plt.axis('off')
plt.show()
```
